## NGINX

### Introduction
* what is NGINX?
  + a web server that serves web content to web browser
  + Used as a reverse proxy, as shown below
    
    + act as a load balancer to forward clients' requests to multiple backend servers
    + send the response from backend servers back to clients
    + clients only send requests to NGINX without managing which backend server to send the requests to
    + this is helpful to scale up the system
  + Used to encrypt requests and responses between clents and servers   
    + makes it easier to set up HTTPS
