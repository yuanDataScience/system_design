## Load Balancer

### Overview
* Load balancer is a machine that runs a reverse proxy software
* The goal is to distribute the requests between multiple servers that host the actual application
* Users don't realize the server behind
* Thousands of clients can issue requests to  servers, load balancers can be between clients and servers to evenly distribute load between servers

### Strategies

#### Round robin
* go through all the server to distribute work loads
* Work loads are evenly distributed
* Avoid overloading one server
* Not efficient if servers have different capacities and resources

#### Performance and resource based load balance
* load balancer performs health checks on 
  + how much traffic a server is handling at any time
  + How long a server is taking to respond to traffic
  + How many resources a server is using
* Load balancer distributes work loads based on the health check results
* Load balancer stops passing requests to low performance or broken servers 
* It evens the loads between servers with better throughput and  latency

#### Path based server selection strategy
* distributes requests according to the path of the requests
* distribute request to different servers according to the path the requests go
  + Payment requests are sent to payment servers
  + Post requests are sent to post servers
* If code of specific functions needs updates, other functions are not affected
* If servers of specific functions are broken, other functions are not affected


#### Weighted round robin
* Distribute work loads according to the resources/capacity of servers

#### Random
* If one  server is overloaded by chance so it can create problems

### Types of load balancer

#### Layer 4 
* transport layer
* has access to TCP r UDP using IP port
* maintain session/connection using session stickiness by assigning an id attribute to a user

#### Layer 7
* application layer
* has access to layer 4 and HTTP headers, cookies and payload
* more complex than layer 4 Load balancer

### Where to place load balancers
* Between clients and servers
* Between servers and databases
* At DNS layer where IP resolving happens
  + Multiple servers deployed, each assigned a unique IP
  + Load balancer is between servers and requests
  + Load balancer resolves IP and balance the requests to servers


### Advantages
* make systems more resilient by monitoring and detecting severs and stop sending requests to unhealthy servers to imporve availability
* improve the scalability of the system by horizonal scaling
* imporve the peformance by distributing requests to severs with less load, and thus avoid overloading and network congestion

### Other key services of load balancer
* health checking: use heartbeat protocol to check healthy and reliability of servers
* TLS termination: decreases the burden of end servers by handling TLS termination with clients. 
  + load balancers will decrypt the code and forward it to end servers and reduce the decryption workload for end servers
  + AWS NLB also has the TLS termination function
* predictive analytics: predict traffic patterns through analytics over traffic through them
* decrease human intervention by decreasing system administration required in handling failures
* service discovery: forward client's requests to appropriate hosting server by inquiring serviced registry
* security: mitigate DDOS at different OSI model (Layers 3, 4 and 7)
